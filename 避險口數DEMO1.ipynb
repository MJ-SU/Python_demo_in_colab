## 步驟二：載入所需套件

```python
import pandas as pd
```

## 步驟三：載入CSV資料


```python
# 計算基準日
calculation_date_str = '2026-03-25'

# 載入調整後股價及報酬率，嘗試使用 'big5' 編碼
df_prices_returns = pd.read_csv('/content/Adjust price20260325.csv', encoding='big5')
print("Adjusted Prices and Returns Data Loaded:")
display(df_prices_returns.head())

# 載入大台指期貨價格，嘗試使用 'big5' 編碼
df_taiex_futures = pd.read_csv('/content/TX202603.csv', encoding='big5')
print("\nTAIEX Futures Data Loaded:")
display(df_taiex_futures.head())

# 載入個別股票的Beta值檔案，嘗試使用 'big5' 編碼
beta_files = {
    '2330': '/content/Beta2330.csv', # 台積電
    '2308': '/content/Beta2308.csv', # 台達電
    '2383': '/content/Beta2383.csv', # 台光電
    '2912': '/content/Beta2912.csv', # 統一超
    '2891': '/content/Beta2891.csv'  # 中信金
}

df_betas = {}
for stock_code, filepath in beta_files.items():
    df_betas[stock_code] = pd.read_csv(filepath, encoding='big5')
    print(f"\nBeta Data for {stock_code} Loaded:")
    display(df_betas[stock_code].head())
```

## 步驟四：計算投資組合總市值

計算投資組合在 `2026年3月25日` 的總市值。

```python
# 投資組合持股組成: {股票代碼: 持股股數} (一張=1000股)
portfolio_composition = {
    '2330': 50 * 1000, # 台積電 (TSMC) - 50張
    '2308': 40 * 1000, # 台達電 (Delta) - 40張
    '2383': 30 * 1000, # 台光電 (Taiming) - 30張
    '2912': 20 * 1000, # 統一超 (Uni-President) - 20張
    '2891': 10 * 1000  # 中信金 (CTBC Financial) - 10張
}

# 將日期欄位轉換為日期時間格式
df_prices_returns['年月日'] = pd.to_datetime(df_prices_returns['年月日'])

# **修正:** 將收盤價(元)從字串轉換為浮點數，並移除逗號
df_prices_returns['收盤價(元)'] = df_prices_returns['收盤價(元)'].astype(str).str.replace(',', '', regex=False).astype(float)

# 過濾出計算日期的股票價格
# 注意：這裡使用.dt.normalize()來確保日期比較只針對日期部分，忽略時間
daily_prices = df_prices_returns[df_prices_returns['年月日'].dt.normalize() == pd.to_datetime(calculation_date_str).normalize()]

portfolio_value = 0
stock_current_prices = {}
print(f"股票 {calculation_date_str} 收盤價與市值計算：")
for stock_code, shares in portfolio_composition.items():
    # 從 '證券代碼' 欄位中提取數字代碼進行匹配，使用正則表達式更健壯
    price_row = daily_prices[daily_prices['證券代碼'].astype(str).str.extract(r'^(\d+)', expand=False) == stock_code]
    if not price_row.empty:
        closing_price = price_row['收盤價(元)'].iloc[0]
        stock_current_prices[stock_code] = closing_price
        stock_market_value = closing_price * shares
        portfolio_value += stock_market_value
        print(f"  {stock_code}: 收盤價 = {closing_price:,.2f} 元, 持股股數 = {shares} 股, 市值 = {stock_market_value:,.2f} 元")
    else:
        print(f"警告：股票 {stock_code} 在 {calculation_date_str} 的價格未找到。")

print(f"\n投資組合總市值 ({calculation_date_str}): {portfolio_value:,.2f} NTD")
```

## 步驟五：處理並計算投資組合的Beta值

從載入的Beta檔案中提取各股票的Beta值，並計算投資組合的加權平均Beta。

```python
# 提取各股票的Beta值
stock_betas = {}
print("\n提取股票Beta值：")
for stock_code, df_beta in df_betas.items():
    beta_value = None
    # 嘗試尋找 '年月日' 或 '日期' 欄位
    beta_date_col = '年月日' if '年月日' in df_beta.columns else ('日期' if '日期' in df_beta.columns else None)
    if beta_date_col:
        df_beta[beta_date_col] = pd.to_datetime(df_beta[beta_date_col])
        # 過濾出計算日期的Beta值
        filtered_beta_df = df_beta[df_beta[beta_date_col].dt.normalize() == pd.to_datetime(calculation_date_str).normalize()]

        if not filtered_beta_df.empty:
            # **修正:** 明確選取 'CAPM_Beta 一年' 的Beta值
            if 'CAPM_Beta 一年' in filtered_beta_df.columns:
                beta_value = filtered_beta_df['CAPM_Beta 一年'].iloc[0]
            else:
                print(f"警告：股票 {stock_code} 在 {calculation_date_str} 的 'CAPM_Beta 一年' 欄位未找到。將預設為 0。")
        else:
            # 如果計算日期沒有資料，嘗試使用檔案中唯一的Beta值（如果存在）
            if len(df_beta) == 1:
                if 'CAPM_Beta 一年' in df_beta.columns:
                    beta_value = df_beta['CAPM_Beta 一年'].iloc[0]
                else:
                    print(f"警告：股票 {stock_code} 的 'CAPM_Beta 一年' 欄位未找到。將預設為 0。")
            else:
                print(f"警告：未能從檔案中找到股票 {stock_code} 在 {calculation_date_str} 的Beta值。請檢查檔案格式或日期。將預設為 0。")

    if beta_value is not None:
        stock_betas[stock_code] = beta_value
        print(f"  {stock_code}: Beta = {beta_value:.4f}")
    else:
        print(f"警告：未能從檔案中找到股票 {stock_code} 在 {calculation_date_str} 的Beta值。請檢查檔案格式或日期。將預設為 0。")
        stock_betas[stock_code] = 0 # 預設為0以避免錯誤

# 計算投資組合的加權平均Beta值
portfolio_beta = 0
print("\n計算投資組合加權Beta值：")

# 檢查 portfolio_value 是否為零，避免 ZeroDivisionError
if portfolio_value == 0:
    print("錯誤：投資組合總市值為零，無法計算加權Beta值。")
else:
    for stock_code, shares in portfolio_composition.items():
        if stock_code in stock_current_prices and stock_code in stock_betas:
            stock_market_value = stock_current_prices[stock_code] * shares
            weight = stock_market_value / portfolio_value
            portfolio_beta += weight * stock_betas[stock_code]
            print(f"  {stock_code}: 權重 = {weight:.4f}, Beta = {stock_betas[stock_code]:.4f}, 加權Beta貢獻 = {(weight * stock_betas[stock_code]):.4f}")
        else:
            print(f"警告：股票 {stock_code} 的市值或Beta值缺失，無法用於投資組合Beta計算。")

print(f"\n投資組合總Beta值: {portfolio_beta:.4f}")
```

## 步驟六：獲取大台指期貨資訊

從期貨資料中取得 `2026年3月` 的大台指期貨收盤價，並定義其合約乘數。

```python
# 將日期欄位轉換為日期時間格式
df_taiex_futures['日期'] = pd.to_datetime(df_taiex_futures['日期'])

# **修正:** 將收盤價從字串轉換為浮點數，並移除逗號
df_taiex_futures['收盤價'] = df_taiex_futures['收盤價'].astype(str).str.replace(',', '', regex=False).astype(float)

# 過濾出計算日期的期貨價格
# 注意：這裡使用.dt.normalize()來確保日期比較只針對日期部分，忽略時間
futures_data = df_taiex_futures[df_taiex_futures['日期'].dt.normalize() == pd.to_datetime(calculation_date_str).normalize()]

taiex_futures_price = 0
if not futures_data.empty:
    taiex_futures_price = futures_data['收盤價'].iloc[0]
    print(f"大台指期貨 {calculation_date_str} 收盤價: {taiex_futures_price:,.2f}")
else:
    # 如果計算日期沒有資料，則使用df_taiex_futures中唯一的日期（如果只有一個）或最近的日期
    if not df_taiex_futures.empty:
        # 找出df_taiex_futures中唯一的日期
        available_dates = df_taiex_futures['日期'].dt.normalize().unique()
        if len(available_dates) == 1:
            actual_futures_date = available_dates[0].strftime('%Y-%m-%d')
            taiex_futures_price = df_taiex_futures['收盤價'].iloc[0]
            print(f"警告：大台指期貨 {calculation_date_str} 的價格未找到。將使用檔案中唯一的日期 {actual_futures_date} 的價格: {taiex_futures_price:,.2f}。")
        else:
            # 如果有多個日期，可以選擇最接近計算日期的日期，這裡簡單使用第一個
            actual_futures_date = available_dates[0].strftime('%Y-%m-%d')
            taiex_futures_price = df_taiex_futures[df_taiex_futures['日期'].dt.normalize() == available_dates[0]]['收盤價'].iloc[0]
            print(f"警告：大台指期貨 {calculation_date_str} 的價格未找到。將使用檔案中第一個可用的日期 {actual_futures_date} 的價格: {taiex_futures_price:,.2f}。")
    else:
        print(f"警告：大台指期貨 {calculation_date_str} 的價格未找到，且期貨資料檔案為空。預設為 17000。")
        taiex_futures_price = 17000 # 預設值，若實際資料缺失請手動調整

# 大台指期貨合約乘數
taiex_futures_multiplier = 200 # 每點200元

print(f"大台指期貨合約乘數: {taiex_futures_multiplier} NTD/點")
```

## 步驟七：計算避險口數

使用以下公式計算所需的避險口數：

$$ 避險口數 = \frac{投資組合市值 \times 投資組合Beta}{期貨價格 \times 期貨合約乘數} $$

```python
# 檢查分母是否為零以避免錯誤
if (taiex_futures_price * taiex_futures_multiplier) != 0:
    hedging_quantity = (portfolio_value * portfolio_beta) / (taiex_futures_price * taiex_futures_multiplier)
    print(f"計算結果：")
    print(f"  投資組合市值: {portfolio_value:,.2f} NTD")
    print(f"  投資組合Beta: {portfolio_beta:.4f}")
    print(f"  大台指期貨價格: {taiex_futures_price:,.2f}")
    print(f"  大台指期貨合約乘數: {taiex_futures_multiplier} NTD/點")
    print(f"\n所需避險口數: {hedging_quantity:.2f} 口")
else:
    print("錯誤：期貨價格或合約乘數為零，或投資組合市值/Beta為零，無法計算避險口數。")
```
