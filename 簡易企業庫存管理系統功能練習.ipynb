# 簡易企業庫存管理系統
專案目的

模擬企業內部的小型庫存管理工具。
功能：

*   新增產品
*   查詢產品
*   更新庫存
*   刪除產品
*   匯出資料

1. 建立資料

```python
import pandas as pd
inventory = pd.DataFrame({
    "ID":[1,2,3],
    "Product":["Laptop","Mouse","Keyboard"],
    "Stock":[10,50,20],
    "Price":[30000,500,1500]
})

inventory
```

2.查詢功能函數

```python
def search_product(name):
    result = inventory[
        inventory["Product"].str.contains(name)
    ]
    return result

while True:
  name = input('輸入想查詢的產品名稱 (輸入 "q" 結束):')
  if name.lower() == 'q':
    print('結束查詢。')
    break

  result = search_product(name)
  if not result.empty:
    print(f'找到相關產品 "{name}":')
    display(result)
    continue_choice = input('要繼續查詢嗎？ (輸入 "y" 繼續，輸入 "n" 結束): ')
    if continue_choice.lower() == 'n':
      print('結束查詢。')
      break
    elif continue_choice.lower() == 'y':
      continue
    else:
      print('無效輸入，將結束查詢。') # Default to ending if input is not 'y' or 'n'
      break
  else:
    print(f'沒有找到產品 "{name}"。請重新輸入。')
```

3.新增產品函數

```python
def add_product(id_val, name_val, stock_val, price_val):
    global inventory

    new_data = {
        "ID": id_val,
        "Product": name_val,
        "Stock": stock_val,
        "Price": price_val
    }

    # Check if ID already exists (this remains in the function as it's a data integrity check, not input type validation)
    if id_val in inventory['ID'].values:
        raise ValueError(f"ID {id_val} 已存在，請使用不同的 ID。")

    inventory.loc[len(inventory)] = new_data

    return inventory

while True:
    try:
        id_input = input('輸入想新增的物品ID (輸入 "q" 結束): ')
        if id_input.lower() == 'q':
            print('結束新增。')
            break
        try:
            id = int(id_input)
        except ValueError:
            raise ValueError("ID 必須是有效的整數。")
        if id < 0:
            raise ValueError("ID 必須是非負整數。")

        name = input('輸入想新增的產品名稱 (輸入 "q" 結束): ')
        if name.lower() == 'q':
            print('結束新增。')
            break
        if not name.strip():
            raise ValueError("產品名稱 必須是非空字串。")

        stock_input = input('輸入想新增的產品庫存 (輸入 "q" 結束): ')
        if stock_input.lower() == 'q':
            print('結束新增。')
            break
        try:
            stock = int(stock_input)
        except ValueError:
            raise ValueError("庫存 必須是有效的整數。")
        if stock < 0:
            raise ValueError("庫存 必須是非負整數。")

        price_input = input('輸入想新增的產品價格 (輸入 "q" 結束): ')
        if price_input.lower() == 'q':
            print('結束新增。')
            break
        try:
            price = int(price_input)
        except ValueError:
            raise ValueError("價格 必須是有效的整數。")
        if price < 0:
            raise ValueError("價格 必須是非負整數。")

        add_product(id, name, stock, price)
        print(f'成功新增產品: ID={id}, Product={name}')
        display(inventory)

        continue_choice = input('要繼續新增產品嗎？ (輸入 "y" 繼續，輸入 "n" 結束): ')
        if continue_choice.lower() == 'n':
            print('結束新增。')
            break
        elif continue_choice.lower() == 'y':
            continue
        else:
            print('無效輸入，將結束新增。')
            break

    except ValueError as e:
        print(f'輸入錯誤: {e}')
        print('請重新輸入或輸入 "q" 結束。')
    except Exception as e:
        print(f'發生未知錯誤: {e}')
        break
```

4.更新庫存函數

```python
def update_stock(id_val, new_stock_val):
    global inventory

    # Check if ID exists before attempting to update
    if id_val not in inventory['ID'].values:
        raise ValueError(f"ID {id_val} 不存在於庫存中，請檢查。")

    inventory.loc[
        inventory["ID"] == id_val,
        "Stock"
    ] = new_stock_val

    return inventory

while True:
    try:
        id_input = input('輸入要更新庫存的產品ID (輸入 "q" 結束): ')
        if id_input.lower() == 'q':
            print('結束更新庫存。')
            break
        try:
            product_id = int(id_input)
        except ValueError:
            raise ValueError("產品ID 必須是有效的整數。")
        if product_id < 0:
            raise ValueError("產品ID 必須是非負整數。")

        new_stock_input = input('輸入新的庫存數量 (輸入 "q" 結束): ')
        if new_stock_input.lower() == 'q':
            print('結束更新庫存。')
            break
        try:
            new_stock = int(new_stock_input)
        except ValueError:
            raise ValueError("新的庫存數量 必須是有效的整數。")
        if new_stock < 0:
            raise ValueError("新的庫存數量 必須是非負整數。")

        update_stock(product_id, new_stock)
        print(f'成功更新產品 ID {product_id} 的庫存為 {new_stock}。')
        display(inventory)

        continue_choice = input('要繼續更新庫存嗎？ (輸入 "y" 繼續，輸入 "n" 結束): ')
        if continue_choice.lower() == 'n':
            print('結束更新庫存。')
            break
        elif continue_choice.lower() == 'y':
            continue
        else:
            print('無效輸入，將結束更新庫存。')
            break

    except ValueError as e:
        print(f'輸入錯誤: {e}')
        print('請重新輸入或輸入 "q" 結束。')
    except Exception as e:
        print(f'發生未知錯誤: {e}')
        break
```

5.刪除產品函數

```python
def delete_product(id_val):
    global inventory

    # Check if ID exists
    if id_val not in inventory['ID'].values:
        raise ValueError(f"ID {id_val} 不存在於庫存中，無法刪除。")

    # Delete the product by ID
    inventory = inventory[inventory['ID'] != id_val].reset_index(drop=True)

    return inventory

while True:
    try:
        id_input = input('輸入要刪除的產品ID (輸入 "q" 結束): ')
        if id_input.lower() == 'q':
            print('結束刪除產品。')
            break

        try:
            product_id = int(id_input)
        except ValueError:
            raise ValueError("產品ID 必須是有效的整數。")
        if product_id < 0:
            raise ValueError("產品ID 必須是非負整數。")

        delete_product(product_id)
        print(f'成功刪除產品 ID {product_id}。')
        display(inventory)

        continue_choice = input('要繼續刪除產品嗎？ (輸入 "y" 繼續，輸入 "n" 結束): ')
        if continue_choice.lower() == 'n':
            print('結束刪除產品。')
            break
        elif continue_choice.lower() == 'y':
            continue
        else:
            print('無效輸入，將結束刪除產品。')
            break

    except ValueError as e:
        print(f'輸入錯誤: {e}')
        print('請重新輸入或輸入 "q" 結束。')
    except Exception as e:
        print(f'發生未知錯誤: {e}')
        break
```

6.匯出為csv檔案

```python
inventory.to_csv(
    "inventory.csv",
    index=False
)
```
