#2.文字資料建立

```python
documents=[
"FDA regulates medical devices.",
"Class I devices have lower risk.",
"Class II devices require special control.",
"Class III devices have highest risk."
]
display(documents)
```

#3.建立Embedding

```python
from sentence_transformers import SentenceTransformer

#使用小型LLM模型all-MiniLM-L6-v2進行Embedding
model=SentenceTransformer(
    "all-MiniLM-L6-v2"
)


embeddings=model.encode(
    documents
)
```

#4.建立向量搜尋

```python
import faiss


index=faiss.IndexFlatL2(
    embeddings.shape[1]
)


index.add(
    embeddings
)
print('向量搜尋已完成')
```

#5.使用者輸入問題

```python
# 準備一些與文件內容相關的問題
potential_questions = [
    "What does the FDA primarily regulate?",
    "What is the risk level of Class I devices?",
    "What are the special requirements for Class II devices?",
    "Which type of device carries the highest risk?"
]

# 顯示可選擇的問題作為選項
print("請選擇一個問題：")
for i, q in enumerate(potential_questions):
    print(f"{i+1}. {q}")

# 讓使用者輸入選擇的數字
while True:
    try:
        choice_index = int(input("請輸入問題的編號 (例如：1, 2, ...): "))
        if 1 <= choice_index <= len(potential_questions):
            question = potential_questions[choice_index - 1]
            break
        else:
            print(f"無效的編號。請輸入1到{len(potential_questions)}之間的數字。")
    except ValueError:
        print("輸入無效。請輸入一個數字。")

print(f"選擇的問題是：\n{question}")

#轉向量
q_embedding=model.encode(
    [question]
)

#向量搜尋
distance, result=index.search(
    q_embedding,
    k=1
)

# 顯示根據使用者問題找到的最相關文件
print("根據選擇的問題，最相關的文件內容是：")
print(documents[result[0][0]])
```
